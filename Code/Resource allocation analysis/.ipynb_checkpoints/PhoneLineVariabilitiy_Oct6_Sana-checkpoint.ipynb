```python
import pandas as pd
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime

import warnings
```

```python
data_path = "/Users/sana/Downloads/STAT390/Data/CAR_-_EP_Flow_Activity_Queue__Agent_Names"
```

```python
folder = Path(data_path)
files = sorted(list(folder.glob("*.csv")) + list(folder.glob("*.xlsx")))
df_main = pd.DataFrame(columns=['Contact Session ID', 'EP Name', 'Flow Name', 'Activity Name', 'Activity Start Timestamp', 
                                'Queue Name', 'Agent Name', 'Termination Reason'])
df_main
```

```python
i=0
for f in files:
    i = i + 1
    if f.suffix.lower() == ".csv":
        df = pd.read_csv(f, header=2, dtype=str, engine="python", skip_blank_lines=False)
    else:  # .xlsx
        df = pd.read_excel(f, sheet_name=0, header=2, dtype=str)
    df_main = pd.concat([df_main, df], ignore_index=True)
    print(i, f.stem, df.shape)
```

```python
df_main["Activity Start Timestamp"] = pd.to_datetime(df_main["Activity Start Timestamp"], errors="coerce")
# Creating a new column 'hour' as it will be useful to visualize peak calling hours
df_main["hour"] = df_main["Activity Start Timestamp"].dt.hour
df_main["weekday"] = df_main["Activity Start Timestamp"].dt.day_name()
```

```python
df_main.info()
df_main.head(10)
```

```python
df_main['EP Name'].unique()
df_main['Termination Reason'].unique()
```

```python
df_filtered = df_main[df_main['EP Name'] != 'Main Number Telephony EP']

# Total calls per EP Name (excluding main)
ep_total_calls = df_filtered['EP Name'].value_counts().reset_index()
ep_total_calls.columns = ['EP Name', 'total_calls']

plt.figure(figsize=(12,6))
sns.barplot(data=ep_total_calls, x='EP Name', y='total_calls')
plt.xticks(rotation=45, ha='right')
plt.title('Total Calls per EP Name (Excluding Main Number)')
plt.tight_layout()
plt.show()

# heatmap
hourly_calls = df_filtered.groupby(['EP Name','hour']).size().reset_index(name='call_count')
hourly_pivot = hourly_calls.pivot(index='EP Name', columns='hour', values='call_count').fillna(0)

plt.figure(figsize=(14,8))
sns.heatmap(hourly_pivot, annot=True, fmt=".0f", cmap='Blues')
plt.title('Hourly Call Volume per EP Name (Excluding Main Number)')
plt.xlabel('Hour of Day')
plt.ylabel('EP Name')
plt.show()

# hourly heamap
weekday_calls = df_filtered.groupby(['EP Name','weekday']).size().reset_index(name='call_count')
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_pivot = weekday_calls.pivot(index='EP Name', columns='weekday', values='call_count').fillna(0)
weekday_pivot = weekday_pivot[weekday_order]

plt.figure(figsize=(14,8))
sns.heatmap(weekday_pivot, annot=True, fmt=".0f", cmap='Greens')
plt.title('Weekday Call Volume per EP Name (Excluding Main Number)')
plt.xlabel('Weekday')
plt.ylabel('EP Name')
plt.show()

# monthly call
monthly_calls = df_filtered.groupby(['EP Name','month']).size().reset_index(name='call_count')

plt.figure(figsize=(14,8))
sns.lineplot(data=monthly_calls, x='month', y='call_count', hue='EP Name', marker='o')
plt.xticks(range(1,13))
plt.title('Monthly Call Volume per EP Name (Excluding Main Number)')
plt.xlabel('Month')
plt.ylabel('Call Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Count calls per EP Name and Termination Reason
termination_counts = df_filtered.groupby(['EP Name', 'Termination Reason']).size().reset_index(name='call_count')

termination_counts['pct'] = termination_counts.groupby('EP Name')['call_count'].transform(lambda x: x / x.sum() * 100)

termination_chart = termination_counts.pivot(index='EP Name', columns='Termination Reason', values='pct').fillna(0)

termination_chart.plot(
    kind='bar',
    stacked=True,
    figsize=(14,6),
    colormap='tab20'
)
plt.title('Percentage of Termination Reasons per EP Name (Excluding Main Number)')
plt.ylabel('Percentage of Calls')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Termination Reason', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```
